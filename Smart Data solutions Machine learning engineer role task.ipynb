```python
import pandas as pd

def combine_csv_files(csv1_path, csv2_path, output_path):
    # Load the datasets
    df1 = pd.read_csv(csv1_path)
    df2 = pd.read_csv(csv2_path)

    # Set the first column (ID) as the index
    df1.set_index(df1.columns[0], inplace=True)
    df2.set_index(df2.columns[0], inplace=True)

    # Perform a left join using the matching IDs
    combined_df = df1.join(df2, on=df1.index, how='left', lsuffix='_df1', rsuffix='_df2')

    # Drop the redundant ID column from df2
    final_df = combined_df.drop(columns=[combined_df.columns[9]])

    # Save the combined dataset to output path
    final_df.to_csv(output_path)

if __name__ == "__main__":
    csv1_path = "C:\\Users\\hp\\Downloads\\MachineLearningEngineerTest\\Java\\resources\\1.csv"
    csv2_path = "C:\\Users\\hp\\Downloads\\MachineLearningEngineerTest\\Java\\resources\\2.csv"
    output_path = "C:\\Users\\hp\\Downloads\\MachineLearningEngineerTest\\Java\\resources\\final_output.csv"

    combine_csv_files(csv1_path, csv2_path, output_path)
```

To visually demonstrate the overlap of IDs between the two files, I plotted a Venn diagram. This helped to clearly see the number of IDs exclusive to each file and the number they shared.
